```python
import pandas
sous_nutrition = pandas.read_csv("sous_nutrition.csv")
population = pandas.read_csv("population.csv")
sous_nutrition.head()
```

```python
sous_nutrition['Valeur'] = pandas.to_numeric(sous_nutrition['Valeur'], errors='coerce')
sous_nutrition.fillna(0, inplace=True)
sous_nutrition.rename(columns={"Valeur": "Sous_nutrition"}, inplace=True)
sous_nutrition['Sous_nutrition'] *= 1000000
sous_nutrition.head()
```

```python
population['Valeur'] *= 1000
population.rename(columns={"Valeur": "Population"}, inplace=True)
population.head()
```

```python
proportion = pandas.merge(population.loc[population['Année'] == 2017, ["Zone", "Population"]], sous_nutrition.loc[sous_nutrition['Année'] == '2016-2018', ["Zone", "Sous_nutrition"]], on='Zone')
proportion.head()
```

```python
pop2017 = proportion['Population'].sum()
sous_nutri2017 = proportion['Sous_nutrition'].sum()
resultat2017 = round((sous_nutri2017/pop2017)*100, 2)
print(f"La proportion de personnes en sous nutrition en 2017 est de :",resultat2017,"%")
```

```python
dispo = pandas.read_csv("dispo_alimentaire.csv")
dispo.fillna(0, inplace=True)
dispo.head()
```

```python
dispo = dispo.merge(population.loc[population['Année'] == 2017, ["Zone", "Population"]], on='Zone')
dispo.head()
```

```python
dispo['dispo_prot'] = dispo['Disponibilité de protéines en quantité (g/personne/jour)'] * dispo['Population'] * 365
print("Quantité de protéines totale disponibles : ",dispo['dispo_prot'].sum(),"g")
```

```python
nb_humain_nourris = round(dispo['dispo_prot'].sum()/(50 * 365), 2)
print("Nombre d'humains pouvant être nourris avec les protéines disponibles :",nb_humain_nourris)
```

```python
prop_humains = round((nb_humain_nourris / population.loc[population['Année'] == 2017, "Population"].sum()) *100, 2)
print("Proportion d'humains pouvant être nourris avec les protéines disponibles :", prop_humains, "%")
```

```python
prod_vegetaux = dispo.loc[dispo['Origine'] == "vegetale"]
print("Quantité de protéines totale disponibles : ",prod_vegetaux['dispo_prot'].sum(),"g")
nb_humain_nourris = round(prod_vegetaux['dispo_prot'].sum()/(50 * 365))
print("Nombre d'humains pouvant être nourris avec les protéines disponibles :",nb_humain_nourris)
prop_humains = round((nb_humain_nourris / population.loc[population['Année'] == 2017, "Population"].sum()) *100, 2)
print("Proportion d'humains pouvant être nourris avec les protéines disponibles :", prop_humains, "%")
```

```python
dispo_int = dispo['Disponibilité intérieure'].sum()
alim_animaux = round((dispo['Aliments pour animaux'].sum()/dispo_int)*100, 2)
pertes = round((dispo['Pertes'].sum()/dispo_int)*100, 2)
nourriture = round((dispo['Nourriture'].sum()/dispo_int)*100, 2)
print("Proportion d'Aliments pour Animaux :",alim_animaux, "%")
print("Proportion de Pertes :",pertes, "%")
print("Proportion de Nourriture :",nourriture, "%")
```

```python
liste_cereales = pandas.read_csv("liste_cereales.csv")
liste_cereales.Produit.unique()
liste_cereales = ['Blé et produits', 'Riz et produits', 'Orge et produits',
       'Maïs et produits', 'Seigle et produits', 'Avoine',
       'Millet et produits', 'Sorgho et produits', 'Céréales, Autres']
prop_animale = dispo.loc[dispo['Produit'].isin(liste_cereales)]
prop_animale = round((prop_animale['Aliments pour animaux'].sum()/prop_animale['Disponibilité intérieure'].sum())*100, 2)
print("La proportion d'alimentation animale est de :", prop_animale, "%")
prop_humaine = dispo.loc[dispo['Produit'].isin(liste_cereales)]
prop_humaine = round((prop_humaine['Nourriture'].sum()/prop_humaine['Disponibilité intérieure'].sum())*100, 2)
print("La proportion d'alimentation pour les humains est de :", prop_humaine, "%")
```

```python
thailande = proportion.loc[proportion['Zone'] == 'Thaïlande']
prop_thai = round((thailande['Sous_nutrition'].sum()/thailande['Population'].sum())*100,2)
print("La proportion de Thailandais en sous nutrition est de :", prop_thai,"%")
```

```python
manioc = dispo.loc[(dispo['Produit'] == "Manioc") & (dispo['Zone'] == 'Thaïlande')]
prop_manioc = round((manioc['Exportations - Quantité'].sum()/manioc['Production'].sum())*100, 2)
print("La proportion de manioc exporté est de : ", prop_manioc,"%")
```

```python
proportion['Proportion'] = round(proportion['Sous_nutrition'] / proportion['Population'] *100, 2)
proportion[['Zone', 'Proportion']].sort_values(by = 'Proportion', ascending=False).head(20)
```
